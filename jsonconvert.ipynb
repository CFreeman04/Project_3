```python
import pandas as pd
```

```python
csv_file = "Cleaned_IncomeData.csv"
cleaned_income_df = pd.read_csv(csv_file, sep=",")
cleaned_income_df
```

```python
cleaned_income_df = cleaned_income_df.drop_duplicates(subset=["State_Name", "City"], ignore_index=True)
cleaned_income_df
```

```python
json_output = "Cleaned_IncomeData.json"
output = cleaned_income_df.to_json(json_output, indent = 4, orient="records")
```
